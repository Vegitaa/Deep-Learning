### 1. Reading a coloured image
By default image is read as BGR in cv2

```python
import cv2
img1 = cv2.imread('/content/CuteCat.jpeg')
img1
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
```

## Converting BGR to RGB

```python

img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2
```

```python
plt.imshow(img2)
```

## Seperate RGB Channels

```python
R, G, B = cv2.split(img2)
```

```python
R
```

```python
G
```

```python
B
```

```python
plt.imshow(R, cmap='Reds')
```

```python
plt.imshow(G, cmap='Greens')
```

```python
plt.imshow(B, cmap='Blues')
```

### Read above image as grayscale


```python
img_gray = cv2.imread('/content/CuteCat.jpeg', cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
plt.imshow(img_gray, cmap='gray')
```
